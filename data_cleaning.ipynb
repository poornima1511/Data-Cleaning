```python
def update_check(ele):
    if pd.isnull(ele):
        return(np.nan)
    elif (ele=='-'):
        return(False)
    else:
        return(True)
c=["job_dissatisfaction","dissatisfaction_with_the_department","physical_work_environment","lack_of_recognition","lack_of_job_security","work_location","employment_conditions","work_life_balance","workload"]

p=tafe_resignation[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]].applymap(update_check)
x=p.any(axis=1,skipna=False)
tafe_resignation["dissatisfied"]=x
tafe_resignation_up=tafe_resignation.copy()
w=dete_resignation[c].applymap(update_check)
dete_resignation["dissatisfied"]=w.any(axis=1,skipna=False)
dete_resignation["dissatisfied"].value_counts(dropna=False)
dete_resignation_up=dete_resignation.copy()
```

```python
dete_resignation_up["institute"]="DETE"
tafe_resignation_up["institute"]="TAFE"
combined=pd.concat([dete_resignation_up,tafe_resignation_up],axis=0,ignore_index=True)
combined.shape
```

```python
combined.info()
```

```python
combined_update=combined.dropna(thresh=500,axis=1)
```

```python
combined_update["institute_service"].value_counts(dropna=False)
```

```python
def institute_Category(element):
    if pd.isnull(element):
        return(np.nan)
    elif int(element) <3:
        return("New: Less than 3 years at a company")
    elif (int(element)>3 and int(element)<6):
        return("Experienced: 3-6 years at a company")
    elif (int(element)>7 and int(element)<10):
        return("Established: 7-10 years at a company")
    elif int(element)>11:
         return("Veteran: 11 or more years at a company")
combined_update["institute_service"]=combined_update["institute_service"].astype(str).str.extract(r"(\d+)")
```

```python
combined_update["institute_Category"]=combined_update["institute_service"].apply(institute_Category)
```
